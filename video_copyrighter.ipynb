```python
import ffmpeg
from PIL import Image
```

```python
video_file = str(input("Enter the the video path : "))
```

```python
video_file
```

```python
logo_file = str(input("Enter the logo path : "))
```

```python
logo_file
```

```python
probe = ffmpeg.probe(video_file)
```

```python
width = probe['streams'][0]['width']
height =probe['streams'][0]['height']
```

```python
img = Image.open(logo_file)
```

```python
img
```

```python
img_width = img.size[0]
img_height = img.size[1]
```

```python
img_height
```

```python
img_width
```

```python
img_resize = img.resize((int(img_width/2), int(img_height/2)))
```

```python
img_resize.save("data/resize_logo.png")
```

```python
in_file = ffmpeg.input(video_file)
overlay_file = ffmpeg.input("data/resize_logo.png")
(
    ffmpeg
    .concat(
        in_file
    )
    .overlay(overlay_file)
    .output('data/out.mp4')
    .run()
)
```
